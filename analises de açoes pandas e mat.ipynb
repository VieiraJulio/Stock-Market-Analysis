```python
import pandas as pd
from cycler import cycler
import matplotlib.pyplot as plt

cores = plt.get_cmap("Pastel2").colors
ciclo_cores = cycler("color", cores)
plt.rc("axes", prop_cycle = ciclo_cores)


ARQUIVO_TOP_10_IBOV = 'top_10_ibov_mai_ago_2023.csv'

df_top_10_ibov = pd.read_csv(ARQUIVO_TOP_10_IBOV)
df_top_10_ibov.head()
```

```python
df_top_10_ibov.info()
```

```python
df_top_10_ibov['Date'] = pd.to_datetime(df_top_10_ibov['Date'] , format='%Y-%m-%d')
df_top_10_ibov.info()
```

```python
df_top_10_ibov.head()
```

```python
df_top_10_ibov.describe()
```

```python
df_top_10_ibov = df_top_10_ibov.set_index('Date',drop=True)

df_top_10_ibov.head()
```

```python
df_top_10_ibov.plot();
```

```python
df_top_10_ibov.plot(figsize=(16, 9), title='Top 10 IBOV', ylabel='Valor em R$')

plt.show()
```

```python
df_top_10_ibov.plot(figsize=(10, 12), title='Top 10 IBOV', subplots=True, layout=(5, 2))
plt.show()
```

```python
df_top_10_ibov.index
```

```python
fig, ax = plt.subplots()

ax.plot(df_top_10_ibov.index, df_top_10_ibov['BBDC4'], label =  'BBDC4')
ax.plot(df_top_10_ibov.index, df_top_10_ibov['ITUB4'], label = 'ITUB4')
ax.plot(df_top_10_ibov.index, df_top_10_ibov['BBAS3'], label = 'BBAS3')
ax.set_ylabel('Variao em milhes de reais')
ax.legend()
ax.set_title('Comparando bancos do top 10 IBOV')

plt.show()
```

```python
variacao_diaria = df_top_10_ibov.pct_change()
variacao_diaria
```

```python
fig, ax = plt.subplots()

ax.plot(variacao_diaria.index, variacao_diaria['BBDC4'], label =  'BBDC4')
ax.plot(variacao_diaria.index, variacao_diaria['ITUB4'], label = 'ITUB4')
ax.plot(variacao_diaria.index, variacao_diaria['BBAS3'], label = 'BBAS3')
ax.legend()
ax.set_ylabel('Variao em %')
ax.set_title('Comparando bancos do top 10 IBOV')

plt.show()
```

```python
df_top_10_ibov.columns.sort_values()
```

```python
colunas = ['BBDC4','ITUB4','BBAS3']
colunas = sorted(colunas)

fig, axs = plt.subplots(nrows = 3, ncols = 1, figsize = (10,10), sharex=True, sharey=True)
for posicao, coluna in enumerate(variacao_diaria[colunas]):
    axs[posicao].plot(variacao_diaria.index, variacao_diaria[coluna], label = coluna, color = cores[posicao])
    axs[posicao].tick_params(axis = 'x', length = 0)
    axs[posicao].legend()
    
fig.suptitle('Comparando bancos do top 10 IBOV', color = 'Darkgray', fontsize = 20, fontweight = 'bold')
fig.supylabel('Variao Percentual', fontsize = 20, fontweight = 'bold')

plt.show()    
```

```python
from itertools import combinations 

bancos = ['BBDC4','ITUB4','BBAS3']

pares_bancos = list(combinations(bancos,2))

fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (12,6))

for posicao, par in enumerate(pares_bancos):
    axs[posicao].scatter(par[0],par[1], data = df_top_10_ibov, color = cores[posicao])
    axs[posicao].set_ylabel(par[0])
    axs[posicao].set_xlabel(par[1])
    axs[posicao].set_title(f'{par[0]} x {par[1]}')
    
fig.suptitle('Analisando correo dos bancos', color = 'Darkgray', fontsize = 20, fontweight = 'bold')

plt.show()
```
